```python
import pandas as pd

# Load the data
df = pd.read_csv('cleaned_data.csv')

# Drop rows with missing churn_risk_score or handle missing data
df.dropna(subset=['churn_risk_score'], inplace=True)
```

```python
df
```

```python
from sklearn.model_selection import train_test_split

# Assuming 'df' is your DataFrame
train, test = train_test_split(df, test_size=0.2, random_state=900)

# 'train' will be 80% of the data
# 'test' will be 20% of the data
```

```python
# Save the train DataFrame to a CSV file
train.to_csv('train_data_2.csv', index=False)
test.to_csv('test_data_2.csv', index=False)
```
