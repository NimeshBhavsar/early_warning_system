# Outlier Detection

```python
# View only numerical columns
df_numerical = df.select_dtypes(include=['number'])

print(df_numerical)
```

```python
import matplotlib.pyplot as plt
# Create a box plot
plt.figure(figsize=(8, 6))
plt.boxplot(df['days_since_last_login'])
plt.title('Box Plot of Days Since Last Login')
plt.ylabel('Days Since Last Login')
plt.xticks([1], ['days_since_last_login'])  # Label x-axis
plt.grid(True)
plt.show()
```

No outliers are detected in days_since_last_login column

```python
# Calculate the mean and standard deviation for each column
means = df_numerical.mean()
std_devs = df_numerical.std()

# Calculate z-scores
z_scores = (df_numerical - means) / std_devs

# Filter z-scores outside the range -1 to 1
z_scores_outside_range = z_scores[(z_scores < -1) | (z_scores > 1)]

print(z_scores_outside_range)
```

```python
def calculate_iqr_and_outliers(df, column_name):
    # Calculate Q1 (25th percentile) and Q3 (75th percentile)
    Q1 = df[column_name].quantile(0.25)
    Q3 = df[column_name].quantile(0.75)
    
    # Calculate IQR
    IQR = Q3 - Q1
    
    # Define outlier bounds
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    
    # Identify outliers
    outliers = df[(df[column_name] < lower_bound) | (df[column_name] > upper_bound)]
    
    # Calculate percentage of outliers
    percentage_outliers = (len(outliers) / len(df)) * 100
    
    return {
        'IQR': IQR,
        'lower_bound': lower_bound,
        'upper_bound': upper_bound,
        'number_of_outliers': len(outliers),
        'percentage_of_outliers': percentage_outliers
    }

# Columns to analyze
columns = ['days_since_last_login', 'avg_time_spent', 'avg_transaction_value', 'avg_frequency_login_days']

# Create a dictionary to store results
results = {}

# Calculate IQR, outliers, and percentage for each column
for col in columns:
    results[col] = calculate_iqr_and_outliers(df, col)

# Convert results to DataFrame for better readability
results_df = pd.DataFrame(results).T

print(results_df)
```

->For avg_time_spent we keep the outliers as the percentage is above 5%.

->For avg_transaction_value I noticed a trend that the outliers were all above 80000 which may mean    that the high spending customer may have a lesser churn risk score.

->For avg_frequency_login_days we keep the outliers as even after deleting the rows with -999 it's showin the same percentage of outliers.
